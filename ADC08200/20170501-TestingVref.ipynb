# Testing acquisition with a changing Vref

on the ADC08200 board

```python
File = ["10MHz/20170501-ADC08_raw_c.csv","10MHz/20170501-ADC08_raw_d.csv","10MHz/20170501-ADC08_raw_e.csv"]
import matplotlib.pyplot as plt
import numpy as np
from ADC08200 import CreateSeries
from scipy.signal import hilbert, chirp
```

## Testing

Plotting the amplified signal from the emulator at different VRerf: see if we can have better resolution. 

```python
NbFiles = len(File)
f, Array = plt.subplots(NbFiles, sharex=True,figsize=(15,5))
              
for i in range(NbFiles):
    Res, t, N = CreateSeries(File[i],1000.0)
    Array[i].plot(t, Res)

plt.show()
```

## Conclusion

Let's keep the range at max! And don't use a settable Vref on the further ADC boards!
