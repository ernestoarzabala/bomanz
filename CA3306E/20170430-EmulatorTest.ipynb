## Testing the acquisition with the emulator

```python
File = "data/10MHz_emul.csv"
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open(File) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
V = [x.strip().split("\t")[1] for x in content] 
V = np.array(V).astype(np.float)*3.3/5.0
t = [x.strip().split("\t")[0] for x in content]

t = np.array(t).astype(np.float)/1000.0
N = float(len(t))
print(N)
print(t[-1])
```

```python
RefT = 6500
plt.plot(t[5500:6500],V[5500:6500])  # plotting 50x 100ns, that's 5µs
plt.ylabel('ADC output')
plt.xlabel("Time (us)")
plt.show()
```

##  Checking the resolution of the ADC

```python
FFT = np.fft.fft(V)
FFT[0]=0
Freqs =  [x/t[-1]*1000.0 for x in range(5000)] # Renormalize FFT
plt.plot(Freqs,abs(FFT[0:5000]))  # plotting 50x 100ns, that's 5µs
plt.ylabel('FFT')
plt.xlabel("Freqency (kHz)")
plt.show()
```
